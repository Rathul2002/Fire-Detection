```python
import cv2
import numpy as np
from ultralytics import YOLO
import winsound
```

```python
model = YOLO('./runs/detect/train2/weights/fire.pt')
```

```python
#To Detect in video

video=cv2.VideoCapture("./Fire.jpg")
while True:
    flag, frame = video.read()
    if not flag:
        break
    results = model.predict(source=frame)
    result=results[0]
    img_np = np.array(result.plot()[:, :, ::-1])
    cv2.imshow("result", cv2.cvtColor(img_np,cv2.COLOR_BGR2RGB))
    no_of_fires=len(result.boxes)
    if no_of_fires>0:
        frequency = 3000 
        duration = 100 
        winsound.Beep(frequency, duration)
    if cv2.waitKey(2) == 27:
        break
video.release()
cv2.destroyAllWindows()
```

```python
#To detect in picture
import matplotlib.pyplot as plt

frame = cv2.imread("./Fire.jpg")
results = model.predict(source=frame)
result = results[0]
img_np = np.array(result.plot())
plt.imshow(cv2.cvtColor(img_np,cv2.COLOR_BGR2RGB))
plt.title("Result")
plt.show()
```
