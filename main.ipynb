```python
from ultralytics import YOLO
```

```python
model = YOLO('./runs/detect/train/weights/last.pt')
```

```python
import cv2
import numpy as np

video=cv2.VideoCapture("./fire1.mp4")
while True:
    flag, frame = video.read()
    if not flag:
        break
    results = model.predict(source=frame)
    result=results[0]
    img_np = np.array(result.plot()[:, :, ::-1])
    cv2.imshow("result", cv2.cvtColor(img_np,cv2.COLOR_BGR2RGB))

    if cv2.waitKey(2) == 27:
        break
video.release()
cv2.destroyAllWindows()
```
